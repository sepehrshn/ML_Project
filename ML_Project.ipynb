```python
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn import svm, metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = shuffle(data)
```

```python
training_data =data.iloc[:629]
```

```python
training_lable=training_data.iloc[:,-1]
```

```python
training_data=training_data.iloc[:,:-1]
```

without set scale of data

```python
svm_clf = svm.SVC(gamma=0.001)
svm_clf.fit(training_data, training_lable)
```

```python
expected = data.iloc[629:,-1]
predicted = svm_clf.predict(data.iloc[629:,:-1])
print(metrics.classification_report(expected, predicted))
```

```python
knn_Model = KNeighborsClassifier(n_neighbors=200)
knn_Model.fit(training_data, training_lable)
```

```python
expected = data.iloc[629:,-1]
peredict_Data = data.iloc[629:,:-1]
predicted = knn_Model.predict(peredict_Data)
print(metrics.classification_report(expected, predicted))
```

with scaler

```python
dataScaler = StandardScaler()
training_data = dataScaler.fit_transform(training_data)
training_data
```

```python
svm_clf = svm.SVC(gamma=0.001)
svm_clf.fit(training_data, training_lable)
```

```python
expected = data.iloc[629:,-1]
predicted = svm_clf.predict(dataScaler.fit_transform(data.iloc[629:,:-1]))
print(metrics.classification_report(expected, predicted))
```

```python
knn_Model = KNeighborsClassifier(n_neighbors=200)
knn_Model.fit(training_data, training_lable)
```

```python
expected = data.iloc[629:,-1]
peredict_Data = dataScaler.fit_transform(data.iloc[629:,:-1])
predicted = knn_Model.predict(peredict_Data)
print(metrics.classification_report(expected, predicted))
```
